```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
L_dist =ctrl.Antecedent(np.arange(0,11,1),'left_distance')
F_dist =ctrl.Antecedent(np.arange(0,11,1),'front_distance')
R_dist =ctrl.Antecedent(np.arange(0,11,1),'right_distance')
L_velo =ctrl.Consequent(np.arange(0,11,1),'left_velocity')
R_velo =ctrl.Consequent(np.arange(0,11,1),'right_velocity')
```

```python
L_dist['near']=fuzz.gaussmf(L_dist.universe,0,2.5)
L_dist['medium']=fuzz.gaussmf(L_dist.universe,5,2.5)
L_dist['far']=fuzz.gaussmf(L_dist.universe,10,2.5)

F_dist['near']=fuzz.gaussmf(F_dist.universe,0,2.5)
F_dist['medium']=fuzz.gaussmf(F_dist.universe,5,2.5)
F_dist['far']=fuzz.gaussmf(F_dist.universe,10,2.5)

R_dist['near']=fuzz.gaussmf(R_dist.universe,0,2.5)
R_dist['medium']=fuzz.gaussmf(R_dist.universe,5,2.5)
R_dist['far']=fuzz.gaussmf(R_dist.universe,10,2.5)

L_velo['slow']=fuzz.gaussmf(L_velo.universe,0,2.5)
L_velo['medium']=fuzz.gaussmf(L_velo.universe,5,2.5)
L_velo['fast']=fuzz.gaussmf(L_velo.universe,10,2.5)

R_velo['slow']=fuzz.gaussmf(R_velo.universe,0,2.5)
R_velo['medium']=fuzz.gaussmf(R_velo.universe,5,2.5)
R_velo['fast']=fuzz.gaussmf(R_velo.universe,10,2.5)

L_dist.view()
F_dist.view()
R_dist.view()
L_velo.view()
R_velo.view()
```

```python
# tạo rule
rule1=ctrl.Rule(L_dist['near'] & F_dist['near'] & R_dist['near'], L_velo['slow'] )
rule2=ctrl.Rule(L_dist['near'] & F_dist['near'] & R_dist['medium'], L_velo['medium'] )
rule3=ctrl.Rule(L_dist['near'] & F_dist['near'] & R_dist['far'], L_velo['fast'] )
rule4=ctrl.Rule(L_dist['near'] & F_dist['medium'] & R_dist['near'], L_velo['slow'] )
rule5=ctrl.Rule(L_dist['near'] & F_dist['medium'] & R_dist['medium'], L_velo['fast'] )
rule6=ctrl.Rule(L_dist['near'] & F_dist['medium'] & R_dist['far'], L_velo['fast'] )
rule7=ctrl.Rule(L_dist['near'] & F_dist['far'] & R_dist['near'], L_velo['slow'] )
rule8=ctrl.Rule(L_dist['near'] & F_dist['far'] & R_dist['medium'], L_velo['fast'] )
rule9=ctrl.Rule(L_dist['near'] & F_dist['far'] & R_dist['far'], L_velo['fast'] )
rule10=ctrl.Rule(L_dist['medium'] & F_dist['near'] & R_dist['near'], L_velo['slow'] )
rule11=ctrl.Rule(L_dist['medium'] & F_dist['near'] & R_dist['medium'], L_velo['slow'] )
rule12=ctrl.Rule(L_dist['medium'] & F_dist['near'] & R_dist['far'], L_velo['fast'] )
rule13=ctrl.Rule(L_dist['medium'] & F_dist['medium'] & R_dist['near'], L_velo['slow'] )
rule14=ctrl.Rule(L_dist['medium'] & F_dist['medium'] & R_dist['medium'], L_velo['medium'] )
rule15=ctrl.Rule(L_dist['medium'] & F_dist['medium'] & R_dist['far'], L_velo['medium'] )
rule16=ctrl.Rule(L_dist['medium'] & F_dist['far'] & R_dist['near'], L_velo['medium'] )
rule17=ctrl.Rule(L_dist['medium'] & F_dist['far'] & R_dist['medium'], L_velo['fast'] )
rule18=ctrl.Rule(L_dist['medium'] & F_dist['far'] & R_dist['far'], L_velo['fast'] )
rule19=ctrl.Rule(L_dist['far'] & F_dist['near'] & R_dist['near'], L_velo['slow'] )
rule20=ctrl.Rule(L_dist['far'] & F_dist['near'] & R_dist['medium'], L_velo['slow'] )
rule21=ctrl.Rule(L_dist['far'] & F_dist['near'] & R_dist['far'], L_velo['slow'] )
rule22=ctrl.Rule(L_dist['far'] & F_dist['medium'] & R_dist['near'], L_velo['slow'] )
rule23=ctrl.Rule(L_dist['far'] & F_dist['medium'] & R_dist['medium'], L_velo['medium'] )
rule24=ctrl.Rule(L_dist['far'] & F_dist['medium'] & R_dist['far'], L_velo['medium'] )
rule25=ctrl.Rule(L_dist['far'] & F_dist['far'] & R_dist['near'], L_velo['slow'] )
rule26=ctrl.Rule(L_dist['far'] & F_dist['far'] & R_dist['medium'], L_velo['medium'] )
rule27=ctrl.Rule(L_dist['far'] & F_dist['far'] & R_dist['far'], L_velo['fast'] )
rule28=ctrl.Rule(L_dist['near'] & F_dist['near'] & R_dist['near'], R_velo['slow'] )
rule29=ctrl.Rule(L_dist['near'] & F_dist['near'] & R_dist['medium'], R_velo['slow'])
rule30=ctrl.Rule(L_dist['near'] & F_dist['near'] & R_dist['far'],  R_velo['slow'])
rule31=ctrl.Rule(L_dist['near'] & F_dist['medium'] & R_dist['near'],  R_velo['slow'])
rule32=ctrl.Rule(L_dist['near'] & F_dist['medium'] & R_dist['medium'],  R_velo['medium'])
rule33=ctrl.Rule(L_dist['near'] & F_dist['medium'] & R_dist['far'],  R_velo['slow'])
rule34=ctrl.Rule(L_dist['near'] & F_dist['far'] & R_dist['near'],  R_velo['slow'])
rule35=ctrl.Rule(L_dist['near'] & F_dist['far'] & R_dist['medium'],  R_velo['medium'])
rule36=ctrl.Rule(L_dist['near'] & F_dist['far'] & R_dist['far'],  R_velo['slow'])
rule37=ctrl.Rule(L_dist['medium'] & F_dist['near'] & R_dist['near'],  R_velo['medium'])
rule38=ctrl.Rule(L_dist['medium'] & F_dist['near'] & R_dist['medium'],  R_velo['slow'])
rule39=ctrl.Rule(L_dist['medium'] & F_dist['near'] & R_dist['far'],  R_velo['slow'])
rule40=ctrl.Rule(L_dist['medium'] & F_dist['medium'] & R_dist['near'],  R_velo['slow'])
rule41=ctrl.Rule(L_dist['medium'] & F_dist['medium'] & R_dist['medium'],  R_velo['medium'])
rule42=ctrl.Rule(L_dist['medium'] & F_dist['medium'] & R_dist['far'],  R_velo['slow'])
rule43=ctrl.Rule(L_dist['medium'] & F_dist['far'] & R_dist['near'],  R_velo['fast'])
rule44=ctrl.Rule(L_dist['medium'] & F_dist['far'] & R_dist['medium'],  R_velo['fast'])
rule45=ctrl.Rule(L_dist['medium'] & F_dist['far'] & R_dist['far'],  R_velo['medium'])
rule46=ctrl.Rule(L_dist['far'] & F_dist['near'] & R_dist['near'],  R_velo['fast'])
rule47=ctrl.Rule(L_dist['far'] & F_dist['near'] & R_dist['medium'],  R_velo['fast'])
rule48=ctrl.Rule(L_dist['far'] & F_dist['near'] & R_dist['far'],  R_velo['slow'])
rule49=ctrl.Rule(L_dist['far'] & F_dist['medium'] & R_dist['near'],  R_velo['fast'])
rule50=ctrl.Rule(L_dist['far'] & F_dist['medium'] & R_dist['medium'], R_velo['fast'])
rule51=ctrl.Rule(L_dist['far'] & F_dist['medium'] & R_dist['far'], R_velo['medium'])
rule52=ctrl.Rule(L_dist['far'] & F_dist['far'] & R_dist['near'],  R_velo['fast'])
rule53=ctrl.Rule(L_dist['far'] & F_dist['far'] & R_dist['medium'], R_velo['fast'])
rule54=ctrl.Rule(L_dist['far'] & F_dist['far'] & R_dist['far'],  R_velo['fast'])
```

```python
navigation=ctrl.ControlSystem([rule1, rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule9,rule10,rule11, rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule19,rule20,rule21, rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule29,rule30,rule31, rule32,rule33,rule34,rule35,rule36,rule37,rule38,rule39,rule39,rule40,rule41, rule42,rule43,rule44,rule45,rule46,rule47,rule48,rule49,rule50,rule51,rule52,rule53])
navi=ctrl.ControlSystemSimulation(navigation)
```

```python
navi.input['left_distance'] =6
navi.input['front_distance']=9
navi.input['right_distance']=6
navi.compute()
print(navi.output['left_velocity'])
print(navi.output['right_velocity'])
L_velo.view(sim=navi)
R_velo.view(sim=navi)
```
